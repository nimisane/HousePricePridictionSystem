```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
```

```python
df = pd.read_csv("dataset/House_Prices.csv")
df.head()
```

```python
df = df.rename({'In which area do you stay in western line in Mumbai?':'Location',
           'What is the carpet area of your current house in square feet?':'Carpet Area',
          'What is the current price of your house in INR?':'Price',
          'How many rooms do you have?':'No of Bedrooms',
          'What facilities are available in your current society?':'Facilities',
          'What all facilities are currently available in your nearby area?':'Near By Facilities'},axis=1)
df.head()
```

```python
facilities = df.iloc[:, 6:]
facilities.head()
```

```python
data = df

def impute_carparking(cols):
    CarParking = cols[0]
    if("Car Parking" in CarParking):
        return 1
    else:
        return 0

def impute_watersupply(cols):
    watersupply = cols[0]
    if("24hr Water Supply" in watersupply):
        return 1
    else:
        return 0
    
def impute_maintenance_staff(cols):
    maintenance_staff = cols[0]
    if("Maintenance Staff" in  maintenance_staff ):
        return 1
    else:
        return 0
    
def impute_Lift(cols):
    lift = cols[0]
    if("Lift" in lift):
        return 1
    else:
        return 0
    
def impute_gaspipeline(cols):
    gas_pipeline = cols[0]
    if("Gas Pipeline" in gas_pipeline ):
        return 1
    else:
        return 0

def impute_gym(cols):
    gym = cols[0]
    if("Gym" in gym):
        return 1
    else:
        return 0

def impute_24hr_security(cols):
    security = cols[0]
    if("24Hr Security" in security):
        return 1
    else:
        return 0


```

```python
data['Car Parking'] = data[['Car Parking']].apply(impute_carparking,axis=1)
data['24hr Water Supply'] = data[['24hr Water Supply']].apply(impute_watersupply,axis=1)
data['Maintenance Staff'] = data[['Maintenance Staff']].apply(impute_maintenance_staff,axis=1)
data['Lift'] = data[['Lift']].apply(impute_Lift,axis=1)
data['Gas Pipeline'] = data[['Gas Pipeline']].apply(impute_gaspipeline,axis=1)
data['Gym'] = data[['Gym']].apply(impute_gym,axis=1)
data['24hr Security'] = data[['24hr Security']].apply(impute_24hr_security,axis=1)
```

```python
def impute_school(cols):
    school = cols[0]
    if("School" in school):
        return 1
    else:
        return 0

def impute_college(cols):
    college = cols[0]
    if("College" in college):
        return 1
    else:
        return 0
    
def impute_medical(cols):
    medical = cols[0]
    if("Medical" in medical):
        return 1
    else:
        return 0
    
def impute_hospitals(cols):
    hospitals = cols[0]
    if("Hospitals" in hospitals):
        return 1
    else:
        return 0
    
def impute_railwayStation(cols):
    railway = cols[0]
    if("Railway Station" in railway):
        return 1
    else:
        return 0

def impute_market(cols):
    market = cols[0]
    if("Market" in market):
        return 1
    else:
        return 0
```

```python
data['School'] = data[['School']].apply(impute_school,axis=1)
data['College'] = data[['College']].apply(impute_college,axis=1)
data['Medical'] = data[['Medical']].apply(impute_medical,axis=1)
data['Hospitals'] = data[['Hospitals']].apply(impute_hospitals,axis=1)
data['Railway Station'] = data[['Railway Station']].apply(impute_railwayStation,axis=1)
data['Market'] = data[['Market']].apply(impute_market,axis=1)
```
